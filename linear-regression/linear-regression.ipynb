```python
import pandas as pd
import numpy as np
import math
import operator
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("onlinedataset.csv")
print(data)
```

```python
X = data['Head Size'].values
Y = data['Brain Weight'].values
print("X ::",X)
print("Y ::",Y)
```

```python
mean_X = np.mean(X)
print("mean_X ::", mean_X)
mean_Y = np.mean(Y)
print("mean_Y ::", mean_Y)
```

```python
length = len(X)

numerator = 0
denominator = 0
for i in range(length):
    numerator += (X[i] - mean_X) * (Y[i] - mean_Y)
    denominator += (X[i] - mean_X)**2

m = numerator/denominator
c = mean_Y - (m*mean_X)

print("m ::", m)
print("c ::", c)
```

```python
max_x = np.max(X)+100
min_x = np.min(Y)-100

x = np.linspace(min_x, max_x, 100)
y = c + m*x


plt.plot(x,y, color = "#58b970", label = "Regression Line")
plt.scatter(X,Y,color = "#ef5423", label="data points")

plt.xlabel("Head Size --------->")
plt.ylabel("Brain Size -------->")
plt.legend()
plt.show()
```

```python
ss_t = 0
ss_r = 0


for i in range(int(length)):
    y_pred = c + m*X[i]
    ss_t += (Y[i] - mean_Y)**2
    ss_r += (Y[i] - y_pred)**2
    
r2 = 1 - (ss_r/ss_t)
print(r2)
    
```
